---

**Ejemplo de la vida real:** Un diagrama de secuencia es como el guion de una obra de teatro: muestra el orden de las interacciones entre los personajes (objetos) a lo largo del tiempo.

```python
# Diagrama de Secuencia UML para el ejemplo Course-Student-Professor

from graphviz import Digraph

dot = Digraph('sequence')

dot.attr(rankdir='LR')

# Participantes
for part in ['Professor', 'Student', 'Course']:
    dot.node(part, part)

# Mensajes
# Professor enseña usando Course y Student
# 1. Professor llama a get_knowledge de Course
# 2. Professor llama a remember de Student

dot.edge('Professor', 'Course', 'get_knowledge()')
dot.edge('Professor', 'Student', 'remember(knowledge)')

# dot.render('uml_sequence_diagram', view=True)
dot
```
