# Ejemplo Integrado: Clases y Diagramas UML en Python

Bienvenido/a. En esta lección verás cómo conectar el código Python con los diagramas UML, integrando clases, diagramas de clases y de secuencia.

## Objetivos
- Relacionar el diseño de clases en Python con su representación en UML.
- Generar y visualizar diagramas de clases y de secuencia a partir de código real.
- Comprender la utilidad de UML para comunicar y documentar sistemas.

---

**Ejemplo de la vida real:** Así como un arquitecto usa planos para explicar una construcción, un desarrollador puede usar UML para explicar el diseño de su software a otros.

```python
# Ejemplo de clases y visualización de diagramas UML en Python

from graphviz import Digraph

# Definición de clases
class Course:
    def get_knowledge(self) -> str:
        return "knowledge"

class Student:
    def remember(self, knowledge: str) -> None:
        print(f"El estudiante ha recordado {knowledge}")

class Professor:
    def __init__(self, student: Student) -> None:
        self.student = student
    def teach(self, course: Course) -> None:
        self.student.remember(course.get_knowledge())

# Instanciación y uso
student = Student()
course = Course()
professor = Professor(student)
professor.teach(course)

# Diagrama de clases UML con graphviz
class_dot = Digraph(comment='Diagrama de Clases')
class_dot.node('Course', 'Course')
class_dot.node('Student', 'Student')
class_dot.node('Professor', 'Professor')
class_dot.edge('Professor', 'Student', label='asociación')
class_dot.edge('Professor', 'Course', label='usa', style='dashed')
class_dot

# Diagrama de secuencia UML con graphviz
seq_dot = Digraph('sequence')
seq_dot.attr(rankdir='LR')
for part in ['Professor', 'Student', 'Course']:
    seq_dot.node(part, part)
seq_dot.edge('Professor', 'Course', 'get_knowledge()')
seq_dot.edge('Professor', 'Student', 'remember(knowledge)')
seq_dot
```
